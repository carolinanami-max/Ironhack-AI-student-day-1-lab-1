```python
def greet_person(name: str, language: str = "Python") -> str:
    """
    Greets a person by name and optionally mentions their favorite programming language.

    Args:
        name (str): The person's name
        language (str, optional): Their favorite programming language. Defaults to "Python".

    Returns:
        str: A personalized greeting message
    """
    # Base greeting message
    greeting = f"Hello, {name}! Welcome to Python programming!"

    # Check if a language is provided AND it's different from the default "Python"
    if language != "Python":
        greeting += f" I see you're interested in {language} too!"

    return greeting

# Test cases for Exercise 1
print("=== Testing Greeting Function ===")
print(greet_person("Alice"))
print(greet_person("Bob", "JavaScript"))
print(greet_person("Charlie", "Python"))
```

```python
def calculate_grade(percentage: float) -> str:
    """
    Converts a percentage score to a letter grade.

    Args:
        percentage (float): The percentage score (0-100)

    Returns:
        str: The letter grade (A, B, C, D, F) or error message
    """
    # Input validation: Check if percentage is within valid range
    if percentage < 0 or percentage > 100:
        return "Invalid percentage. Please enter a value between 0 and 100."

    # Convert percentage to letter grade based on grading scale
    # Conditions are checked in order from highest to lowest
    if percentage >= 90:
        return "A"
    elif percentage >= 80:
        return "B"
    elif percentage >= 70:
        return "C"
    elif percentage >= 60:
        return "D"
    else:
        return "F"

# Test cases for Exercise 2
print("\n=== Testing Grade Calculator ===")
test_scores = [95, 85, 75, 65, 55, 90, 60]
for score in test_scores:
    grade = calculate_grade(score)
    print(f"Score: {score}% -> Grade: {grade}")

# Test edge cases
print("\n=== Testing Edge Cases ===")
print(f"Score: 100% -> Grade: {calculate_grade(100)}")
print(f"Score: 0% -> Grade: {calculate_grade(0)}")
print(f"Score: 59.9% -> Grade: {calculate_grade(59.9)}")
print(f"Score: -5% -> Grade: {calculate_grade(-5)}")
print(f"Score: 105% -> Grade: {calculate_grade(105)}")
```

## Reflection Questions

**1. What is the main advantage of using functions instead of writing the same code multiple times?**

Functions promote code reusability, reduce duplication, and make code more organized and maintainable. If you need to change the logic (like adjusting grade thresholds), you only need to update it in one place.

**2. In Exercise 1, why did we use a default parameter value for language?**

We used a default parameter to make the function more flexible. Users can call it with just a name, and it will use "Python" as the default language. This provides sensible default behavior while still allowing customization.

**3. In Exercise 2, why is it important to check the conditions in order from highest to lowest (90, then 80, then 70, etc.)?**

Checking conditions from highest to lowest ensures that we match the correct grade bracket. If we checked 60+ first, a score of 95 would incorrectly get a "D" instead of "A" because it satisfies the first condition it encounters.
